```python
import numpy as np

line = np.loadtxt('20', skiprows=4)
```

```python
from math import sin, cos, tan, pi
x = 5728374.726
y = 4210198.192
# на входных данных должен быть сдвиг на 500000.00
def From_Gauss_to_Geo(x, y, zone_number):
    L0 = 6*zone_number - 3
    beta = (x/6367558.4969)
    bx = beta + (50221746 + (293622 + (2350 + 22*(cos(beta)**2))*(cos(beta)**2))*(cos(beta)**2))*0.0000000001*sin(beta)*cos(beta)

    nx = 6399698.902 - (21562.267 - (108.973 - 0.612 * (cos(bx)**2))*(cos(bx)**2))*(cos(bx)**2)       
    z = (y - 1000000*zone_number) / (nx * cos(bx))
    b2 = (0.5 + 0.003369 * (cos(bx)**2)) * sin(bx) * cos(bx)

    b3 = 0.333333 - (0.166667 - 0.001123 * (cos(bx)**2)) * (cos(bx)**2)
    b4 = 0.25 + (0.16161 + 0.00562*(cos(bx)**2)) * (cos(bx)**2)
    
    b5 = 0.2 - (0.1667 - 0.0088 * (cos(bx)**2))  * (cos(bx)**2)
    L1 = (1 - (b3 - b5 * (z**2))*(z**2))*z * 180/pi
    L = L0 + L1                 
    B = (bx - (1 - (b4 - 0.12*(z**2))*(z**2))*(z**2)*b2) * 180/pi
    # Широта, долгота
    return B, L
  
```

```python

# Zone number is written out manually

my_str = ''

for i in range(0, len(line[:,0])):  
    a = (line[i, 1] + 850000000) / 100
    b = line[i, 2] / 100
    long, lat = From_Gauss_to_Geo(b, a, 9)
    c_line[i, 0] = line[i, 0]
    c_line[i, 1] = (line[i, 1] + 900000000)/100
    c_line[i, 2] = line[i, 2] / 100
    c_line[i, 3] = lat
    c_line[i, 4] = long
    
    my_str = my_str +  str(long) + ',' + str(lat)  + ',' 
    
print(my_str)    
```
